### Task 1: Visualizing Gender-wise Population Distribution in India Using Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
df_men = pd.read_csv("Population_men.csv", skiprows=4)
print("Men Population Dataset:")
display(df_men)
```

```python
df_women = pd.read_csv("Population_female.csv", skiprows=4)
print("\nWomen Population Dataset:")
display(df_women)
```

```python
df_total = pd.read_csv("Population_total.csv", skiprows=4)
print("\nTotal Population Dataset:")
display(df_total)
```

```python
# Check the columns to confirm the structure
print("Men dataset columns:\n", df_men.columns)
print("\nWomen dataset columns:\n", df_women.columns)
print("\nTotal dataset columns:\n", df_total.columns)
```

```python
# Filter data for India only
india_men = df_men[df_men['Country Name'] == 'India']
india_women = df_women[df_women['Country Name'] == 'India']
india_total = df_total[df_total['Country Name'] == 'India']

# Display the filtered results
print("India - Male Population Data:")
display(india_men)

print("\nIndia - Female Population Data:")
display(india_women)

print("\nIndia - Total Population Data:")
display(india_total)
```

```python
# We'll use the most recent non-null year (e.g., 2022)
# Get the last year column with valid data
year_columns = india_men.columns[4:]  # From 1960 onwards

# Check the most recent year with non-null data
latest_year = year_columns[::-1][india_men[year_columns[::-1]].notnull().any()].values[0]
print("Latest year with data available:", latest_year)
```

```python
# Extract population numbers for the latest year
male_population = india_men[latest_year].values[0]
female_population = india_women[latest_year].values[0]
total_population = india_total[latest_year].values[0]

# Print values
print(f"Male Population in {latest_year}: {male_population:,}")
print(f"Female Population in {latest_year}: {female_population:,}")
print(f"Total Population in {latest_year}: {total_population:,}")
```

```python
# Create a dataframe for plotting
gender_data = pd.DataFrame({
    'Gender': ['Male', 'Female'],
    'Population': [male_population, female_population]
})

# Display the dataframe
display(gender_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot for male and female population
plt.figure(figsize=(8, 6))
sns.barplot(data=gender_data, x='Gender', y='Population', hue='Gender',
            palette=['#1f77b4', '#ff69b4'], legend=False)
plt.title('Male vs Female Population in India (2022)', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Set 'Gender' as hue and remove legend to avoid duplication
sns.barplot(data=gender_data, x='Gender', y='Population', hue='Gender',
            palette=['#1f77b4', '#ff69b4'], legend=False)

plt.title(f'Gender-wise Population Distribution in India ({latest_year})', fontsize=14, weight='bold')
plt.ylabel('Population')
plt.xlabel('Gender')
plt.tight_layout()

# Save the plot as a PNG file
plt.savefig('India_gender_population_bar_chart.png')
plt.show()
```
